```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_csv('/content/mtcars (1).csv')
```

```python
#x and y values for plotting a line plot
x=[-3,5,7]
y=[10,2,5]
```

```python
#Line plot
fig =plt.figure(figsize= (20,3))
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(-4,8)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sales Comparision',size=20,y=1)
```

```python
#Save image & specify the resolution(dpi-dots per inch)
fig.savefig('example.png',dpi=300,bbox_inches='tight')
```

```python
#Subplots
#fig, ax=plt.subplots(nrows,ncols)
fig,ax=plt.subplots(2,4)
```

```python
x =[2,4,6]
x1 =[4,8,12]
y=[3,6,9]
y1=[4,5,8]
```

```python
fig,axes=plt.subplots(2,2)
axes[0,0].plot(x,y)
axes[0,1].plot(x,y1)
axes[1,0].plot(x1,y)
axes[1,1].plot(x1,y1)
```

```python
#Crosstable
pd.crosstab(mtcars.gear,mtcars.cyl)
```

```python
pd.crosstab(mtcars.gear,mtcars.cyl).plot(kind='bar')
```

```python
#Bar Graph
plt.bar(mtcars.gear,mtcars.cyl)
plt.xlim(0,10)
plt.ylim(0,10)
plt.title('Bar Graph')
plt.xlabel('Gear')
plt.ylabel('Cyl')
```

```python
#count for each type of gear
mtcars['gear'].value_counts()
```

```python
#Pie Chart
mtcars['gear'].value_counts().plot(kind='pie')
```

```python
#Scatter plots
plt.scatter(mtcars.mpg,mtcars.qsec)
plt.xlabel('MPG')
plt.ylabel('qsec')
```

```python
#Histogram
plt.hist(mtcars['mpg'])
```

```python
plt.hist(mtcars.mpg, edgecolor='black',facecolor='pink',bins=5)
plt.title('Histogram -MPG')
```

```python
#Box plots
plt.boxplot(mtcars.mpg,vert = True)
plt.title('Boxplot-MPG')
```

```python
plt.boxplot(mtcars.mpg,vert = False)
plt.title('Boxplot-MPG')
```

```python
#Violin plot
plt.violinplot(mtcars.mpg)
```
